### Program Name : Payroll Generator

```python
def calculate_regular_pay(hours_worked, pay_rate):
    '''To Calculate Regular Gross Pay'''
    regular_pay = hours_worked * pay_rate
    return regular_pay    
```

```python
def calculate_overtime_pay(hours_worked, pay_rate):
    '''To calculate Overtime Pay'''
    OVERTIME_MULTIPLIER = 1.5
    MAX_REG_HOURS = 40.00
    overtime_hours_worked=0
    overtime_pay_rate=0
    overtime_pay=0
    
    if(hours_worked > MAX_REG_HOURS):
        overtime_hours_worked = hours_worked - MAX_REG_HOURS
        overtime_pay_rate = OVERTIME_MULTIPLIER * pay_rate
        overtime_pay = overtime_hours_worked * pay_rate * OVERTIME_MULTIPLIER
        
    return overtime_hours_worked, overtime_pay_rate, overtime_pay
```

```python
def calculate_total_gross_pay(regular_pay, overtime_pay):
    '''To calculate the total gross pay'''
    total_gross_pay = regular_pay + overtime_pay
    return total_gross_pay
```

```python
def calculate_tax_withholding(total_gross_pay, pay_rate):
    '''To Calculate Tax Withheld'''
    TAX_RATE_BELOW_12 = 10
    TAX_RATE_BELOW_20 = 15
    TAX_RATE_ABOVE_20 = 20

    if(pay_rate < 12.00):
        tax_rate = TAX_RATE_BELOW_12
    elif(pay_rate >= 12.00 and pay_rate < 20.00):
        tax_rate = TAX_RATE_BELOW_20
    elif(pay_rate >= 20.00):
        tax_rate = TAX_RATE_ABOVE_20
    else:
        print("Tax rate cannot be determined")
    
    tax_withholding = total_gross_pay * (tax_rate/100)
    return tax_withholding
```

```python
def calculate_ded_withholding(total_gross_pay):
    '''To calculate Deductions Withheld'''
    DEDUCTION_RATE_LOW = 2
    DEDUCTION_RATE_HIGH = 5
    DEDUCTION_TRHRESHOLD = 800

    if(total_gross_pay < DEDUCTION_TRHRESHOLD):
        ded_rate = DEDUCTION_RATE_LOW
    elif(total_gross_pay >= DEDUCTION_TRHRESHOLD):
        ded_rate = DEDUCTION_RATE_HIGH
    else:
        print("Deduction rate cannot be determined")
        
    ded_withholding = total_gross_pay * (ded_rate/100)
    return ded_withholding
```

```python
def calculate_net_pay(total_gross_pay, tax_withholding, ded_withholding):
    '''To Calculate the net pay'''
    net_pay = total_gross_pay - tax_withholding - ded_withholding
    return net_pay
```

```python
def display_results(employee_name, hours_worked, pay_rate, regular_pay, 
                    overtime_hours, overtime_pay_rate, overtime_pay, 
                    total_gross_pay, tax_withholding, ded_withholding, net_pay, payroll_csv_output_file):
    '''To Display results in csv and console'''
    print("-"*45)
    print(f"Employee Name: \t\t\t{employee_name}")
    print(f"Regular Hours worked: \t\t {format(hours_worked, '8.2f')}")
    print(f"Hourly Pay Rate: \t\t${format(pay_rate, '8.2f')}")
    print(f"Regular Gross Pay: \t\t${format(regular_pay, '8.2f')}")
    print(f"Overtime Hours Worked: \t\t {format(overtime_hours, '8.2f')}")
    print(f"Overtime Pay Rate: \t\t${format(overtime_pay_rate, '8.2f')}")
    print(f"Overtime Gross Pay: \t\t${format(overtime_pay, '8.2f')}")
    print(f"Total Gross Pay: \t\t${format(total_gross_pay, '8.2f')}")
    print(f"Taxes Withheld: \t\t${format(tax_withholding, '8.2f')}")
    print(f"Deductions Withheld: \t\t${format(ded_withholding,'8.2f')}")
    print(f"Net Pay: \t\t\t${format(net_pay, '8.2f')}")
    print("-"*45)
    print("\n")
    
    payroll_output_csv_format = f"{employee_name},{hours_worked},{pay_rate},{regular_pay},{overtime_hours},{overtime_pay_rate},{overtime_pay},{total_gross_pay},{tax_withholding},{ded_withholding},{net_pay}\n"
    payroll_csv_output_file.write(payroll_output_csv_format)
```

```python
# Your program code should begin in this cell, starting on line 3

def main():
    '''This is the main function of this program'''
    #INITIALIZATION
    FILE_PATH = r'\USERS\CIS5357\EXAMS'
    payroll_input_file = open(FILE_PATH + r'\Payroll_input.csv', 'r')
    payroll_csv_output_file = open(FILE_PATH + r'\Payroll_output.csv', 'w')
    employee_count = 0
    SINGLE_EMPLOYEE_RECD = 3
    EOF = 'EOF'

    employee_name_recd = payroll_input_file.readline().rstrip('\n')

    #PROCESS
    while employee_name_recd != 'EOF':
        for employee in range(1, SINGLE_EMPLOYEE_RECD, 1):
            employee_name = employee_name_recd
            employee_recd_list = payroll_input_file.readline().rstrip('\n')
            if(employee == 1):
                hours_worked = employee_recd_list
                hours_worked = float(hours_worked)
            else:
                pay_rate = employee_recd_list
                pay_rate = float(pay_rate)

        regular_gross_pay = calculate_regular_pay(hours_worked, pay_rate)
        overtime_hours_worked, overtime_pay_rate, overtime_pay = calculate_overtime_pay(hours_worked, pay_rate)
        total_gross_pay = calculate_total_gross_pay(regular_gross_pay, overtime_pay)
        tax_withholding = calculate_tax_withholding(total_gross_pay, pay_rate)
        ded_withholding = calculate_ded_withholding(total_gross_pay)
        net_pay = calculate_net_pay(total_gross_pay, tax_withholding, ded_withholding)
        display_results(employee_name, hours_worked, pay_rate, regular_gross_pay, overtime_hours_worked, overtime_pay_rate, 
                        overtime_pay, total_gross_pay, tax_withholding, ded_withholding, net_pay, payroll_csv_output_file)

        employee_name_recd = payroll_input_file.readline().rstrip('\n')
        employee_count += 1
    
    #FINALIZATION
    payroll_input_file.close()
    payroll_csv_output_file.close()

    print(f"Completed Payroll Processing for {employee_count} employees")
    print(f"{employee_count} employee records were written to output file")
```

```python
# Calling the main method

main()
```
